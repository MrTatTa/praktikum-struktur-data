```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}


# Menampilkan graf
def print_graph(graph):

    for node in graph:
        print(f"{node} -> {graph[node]}")


print_graph(graph)
```

```python
from collections import deque


def bfs(graph, start):

    visited = set()

    queue = deque([start])

    while queue:

        vertex = queue.popleft()

        if vertex not in visited:

            print(vertex, end=" ")

            visited.add(vertex)

            queue.extend(
                neighbor for neighbor in graph[vertex]
                if neighbor not in visited
            )


# Contoh pemanggilan
bfs(graph, 'A')
```

```python
# Lengkapi fungsi untuk menghitung jumlah tetangga dari tiap simpul dalam graf

def count_neighbors(graph):

    result = {}

    for node in graph:

        # Hitung jumlah tetangga
        result[node] = len(graph[node])

    return result


# Contoh penggunaan
hasil = count_neighbors(graph)

for node, jumlah in hasil.items():
    print(f"{node} memiliki {jumlah} tetangga")
```
