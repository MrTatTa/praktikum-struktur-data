```python
print("=" * 50)
print("STUDI KASUS 1 - Nilai Min & Max Mahasiswa")
print("=" * 50)

class pair:
    def __init__(self):
        self.min = 0
        self.max = 0

def getMinMax(arr: list, n: int) -> pair:
    minmax = pair()

    # If there is only one element then return it as min and max both
    if n == 1:
        minmax.max = arr[0]
        minmax.min = arr[0]
        return minmax

    # If there are more than one elements, then initialize min and max
    if arr[0] > arr[1]:
        minmax.max = arr[0]
        minmax.min = arr[1]
    else:
        minmax.max = arr[1]
        minmax.min = arr[0]

    for i in range(2, n):
        if arr[i] > minmax.max:
            minmax.max = arr[i]
        elif arr[i] < minmax.min:
            minmax.min = arr[i]

    return minmax

# Driver Code - 40 nilai mahasiswa
if __name__ == "__main__":
    import random
    random.seed(42)
    arr = [random.randint(60, 100) for _ in range(36)]
    print("Nilai 36 Mahasiswa:")
    print(arr)
    arr_size = len(arr)
    minmax = getMinMax(arr, arr_size)
    print("Minimum element is", minmax.min)
    print("Maximum element is", minmax.max)

print()
```

```python
print("=" * 50)
print("STUDI KASUS 2 - Angka Hilang & Terulang")
print("=" * 50)

def printTwoElements(arr):
    n = len(arr)
    temp = [0] * n  # Creating temp array of size n with initial values as 0.
    angkaterulang = -1
    angkahilang = -1

    for i in range(n):
        temp[arr[i] - 1] += 1
        if temp[arr[i] - 1] > 1:
            angkaterulang = arr[i]
    for i in range(n):
        if temp[i] == 0:
            angkahilang = i + 1
            break

    print("angka yanng terulang adalah", angkaterulang, ".")
    print("angka yang hilang adalah", angkahilang, ".")

# Test dengan contoh dari modul
arr = [7, 3, 4, 5, 5, 6, 2]
printTwoElements(arr)

# Test dengan contoh lain
print()
arr2 = [3, 1, 3]
print("Input:", arr2)
printTwoElements(arr2)

print()
arr3 = [4, 3, 6, 2, 1, 1]
print("Input:", arr3)
printTwoElements(arr3)

print()
```

```python
print("=" * 50)
print("STUDI KASUS 3 - Pasangan Elemen dengan Jumlah X")
print("=" * 50)

def findPairWithSum(arr, X):
    n = len(arr)
    # Gunakan set untuk melacak elemen yang sudah dilihat
    seen = set()
    pairs = []

    for num in arr:
        complement = X - num
        if complement in seen:
            pairs.append((complement, num))
        seen.add(num)

    if pairs:
        print(f"true")
        for p in pairs:
            print(f"Explanation: Terdapat Pasangan {p} dengan jumlah {X}")
    else:
        print(f"false")
        print(f"Explanation: Tidak ada pasangan dengan jumlah {X}.")

# Test Case 1
arr1 = [16, 13, 6, 4, 5, 10]
X1 = 26
print(f"Input: arr[] = {arr1}, X = {X1}")
findPairWithSum(arr1, X1)

print()

# Test Case 2
arr2 = [15, 13, 26, 37, 19, 10]
X2 = 35
print(f"Input: arr[] = {arr2}, X = {X2}")
findPairWithSum(arr2, X2)

print()

# Test Case 3
arr3 = [11, 15, 26, 38, 9, 10]
X3 = 45
print(f"Input: arr[] = {arr3}, X = {X3}")
findPairWithSum(arr3, X3)
```
