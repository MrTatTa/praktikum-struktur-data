```python
# Membuat stack kosong
stack = []

# Push (menambah elemen)
stack.append('a')
stack.append('b')
stack.append('c')

print("Stack setelah push:", stack)

# Pop (menghapus elemen paling atas)
stack.pop()

print("Stack setelah pop:", stack)

# Melihat elemen paling atas
print("Elemen teratas:", stack[-1])

# Mengecek apakah stack kosong
print("Apakah stack kosong?", len(stack) == 0)

# Mengetahui ukuran stack
print("Ukuran stack:", len(stack))
```

```python
from collections import deque

# Membuat stack
stack = deque()

# Push
stack.append('a')
stack.append('b')
stack.append('c')

print("Stack setelah push:", stack)

# Pop
stack.pop()

print("Stack setelah pop:", stack)

# Elemen teratas
print("Elemen teratas:", stack[-1])

# Ukuran stack
print("Ukuran stack:", len(stack))

# Cek kosong
print("Apakah stack kosong?", len(stack) == 0)
```

```python
from queue import LifoQueue

# Membuat stack
stack = LifoQueue(maxsize=5)

# Push
stack.put('a')
stack.put('b')
stack.put('c')

print("Ukuran stack:", stack.qsize())

# Pop
print("Pop:", stack.get())
print("Pop:", stack.get())

print("Ukuran stack sekarang:", stack.qsize())

# Cek kosong
print("Apakah kosong?", stack.empty())
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self):
        self.head = None
        self.size = 0

    def isEmpty(self):
        return self.size == 0

    def push(self, value):
        node = Node(value)
        node.next = self.head
        self.head = node
        self.size += 1

    def pop(self):
        if self.isEmpty():
            print("Stack kosong")
            return None
        removed = self.head.data
        self.head = self.head.next
        self.size -= 1
        return removed

    def peek(self):
        if self.isEmpty():
            return None
        return self.head.data

    def getSize(self):
        return self.size


# Contoh penggunaan
s = Stack()

s.push(10)
s.push(20)
s.push(30)

print("Elemen teratas:", s.peek())
print("Pop:", s.pop())
print("Ukuran stack:", s.getSize())
```
